### Agente

```python
class Agente_grid:
    """Agent in a especific location in a 51x51 discrete grid 
        id: guarda un número único para cada agente
        tipo: un número para efectos de graficación
        existence: booleano que asegura que el agente exista en el mundo, ya que este puede morir o solaparse con otros.
        posicion: np.array (1x2) la posición en ese momento del agente
        vision: entero entre 1 y 2, caracteristica intrinseca de cada agente que le permite tener mayor campo de acción
        vecindad: diccionario donde se guardan los agentes vecinos con los que podría interactuar
        posicion_deseada: posición a la que el agente irá despues de tomar una desición basada en la información recopilada
        max_azucar: azucar máxima que puede conseguir para efectos de toma de decisiones
        azucar_deseada: entero que representa la azucar espeficifica por la que el agente se dirige. 
    """
    
    def __init__(self, id ,tipo = 2, x = None, y = None, vision = None , energy = None, metabolism = None):
        self.id = id
        self.tipo = tipo
        self.sex = choice(['M','F'])
        if metabolism == None:
            self.metabolism = randint(1,3)
        else:
            self.metabolism = metabolism
        if energy == None:
            self.energy = 10
        else:
            self.energy = energy
        self.existence = False
        self.objective = None
        self.vecindad = {'azucar' : [], #Dictionary to store the information of the things nearby
                         'sugar_id' : [],
                         'vecinos' : [],
                         'empty_spaces': []}
        if x == None or y == None:
            self.posicion = np.array([randint(0,50), randint(0,50)])
        else:
            self.posicion = np.array([x,y])
        if vision == None:
            self.vision = randint(1,2)
        else:
            self.vision = vision
        
                           
        
        
    def ver(self, mundo , show = False):
        """
        Basado en el rango de visión, el agente "ve" sus alrededores en busca de azucar y otros agentes.
        Los 4 if's actuan cuando el agente está en alguna de las orillas del mapa evitando salirnos del arreglo.
        """
        if self.posicion[0] - self.vision < 0:
            x_min = 0
        else:
            x_min = self.posicion[0]-self.vision
            
        if  self.posicion[0] + self.vision + 1 > 51:
            x_max = 51
        else:
            x_max = self.posicion[0] + self.vision + 1
            
        if self.posicion[1] - self.vision < 0:
            y_min = 0
        else:
            y_min = self.posicion[1] - self.vision
            
        if self.posicion[1] + self.vision + 1 > 51:
            y_max = 51
        else:
            y_max =self.posicion[1] + self.vision + 1
        if show == True:
            print mundo.grid[x_min : x_max , y_min : y_max]  #Linea para pruebas ELIMINAR
        return np.array([x_min,x_max,y_min,y_max])

    def explorar(self, mundo):
        """
        Recorre el campo de acción del agente basado en su rango de visión
        Identifica y guarda en un diccionario cada tipo de elemento que esté cerca de el para después tomar una desición.
        Como la vecindad cambia despues de cada iteración (o movimiento del agente), limpiamos en cada paso el diccionario de vecindad
        """
        lim = self.ver(mundo)
        self.vecindad['vecinos'] = []
        self.vecindad['azucar'] = []
        self.vecindad['sugar_id'] = []
        self.vecindad['empty_spaces'] = []
        #Clasify the whole neighborhood
        for i in xrange(lim[0],lim[1]): #Renglones
            for j in xrange(lim[2],lim[3]): #Columnas 
                if mundo.grid[i,j] == 0:
                    self.vecindad['empty_spaces'].append(np.array([i,j]))
                elif mundo.grid[i,j] == 1:
                    self.vecindad['azucar'].append(np.array([i,j]))
                elif mundo.grid[i,j] == 2: 
                    self.vecindad['vecinos'].append(np.array([i,j]))
        #Find the sugars ID of the neighbors             
        for i in xrange(len(self.vecindad['azucar'])):
            for j in xrange(len(mundo.sugars_list)):
                if np.all(self.vecindad['azucar'][i] == mundo.sugars_list[j].posicion):
                    self.vecindad['sugar_id'].append(j)
                    break
                
                    
    def decidir(self, mundo): 
        """
        Given that the agent already knows what are its surroundings it tries to maximize its yield doing a cost benefit analysis.
        The agent then moves to the best choice availabe. If there is no good choice, it'll simply move randomly.
        benefit = amount of energy units it'll recive from the sugar. Each sugar gives 1 energy
        cost = amount of energy it'll consume to get there given it's metabolism.
        """
        self.explorar(mundo)
        bc =  -6 #Benefit - Cost
        for i in self.vecindad['sugar_id']:
            if self.benefit(mundo, i) - self.cost(mundo, i) > bc:
                bc = self.benefit(mundo, i) - self.cost(mundo, i)
                self.objective = i
        if bc <= -6:
            self.objective = None
    
    def moverse(self, mundo):
        """
        Como se mueven los agentes.
        Tres tipos de movimiento:
        1. Si no tiene una posición deseada, se mueve aleatoreamente a un espacio vacio.
        2. Si se quiere mover a una casilla adyacente se mueve y el azucar se apaga mientras el agente este sobre el azucar.
        """
        self.decidir(mundo) 
        if self.objective == None:
            random_choice = choice(self.vecindad['empty_spaces'])
            self.energy -= self.metabolism * self.calculate_distance(random_choice)
            self.posicion = random_choice
        else:
            self.energy += self.benefit(mundo, self.objective)  - self.cost(mundo, self.objective)
            self.posicion = mundo.sugars_list[self.objective].posicion
            mundo.sugars_list[self.objective].existence = False #Removes the sugar from existance.
    
    def calculate_distance(self, new_position):
        """
        Claculates the discreete distance between its own position and a new one. Since the particles can move in diagonal,
        we're assuming a diagonal move to be equal to one.
        """
        return int(numpy.linalg.norm(self.posicion - new_position))
    
    def benefit(self, mundo, x):
        return mundo.sugars_list[x].cantidad
    
    def cost(self, mundo, x):
        return self.metabolism * self.calculate_distance(mundo.sugars_list[x].posicion)
    
    def state(self, mundo, full_state = False):
        """
        Handy function to review an specific agent and it's caracteristics
        full_state: boolean that chooses he amount of info that displays.
        """
        if full_state == True:
            print "Hello, I'm a type %s agent with ID: %s" % (self.tipo, self.id)
            print "My stats are as follow:" 
            print "\t Sex: %s" % self.sex 
            print "\t Metabolism: %s" % self.metabolism
            print "\t Vision: %s" % self.vision
            print "My current state is: "
            print "\t Energy: %s" % self.energy
            print "\t Position: %s" % self.posicion
            print "\t Surroundings: "
            self.ver(mundo, show = True)
        else:
            print "\t Energy: %s" % self.energy
            print "\t Position: %s" % self.posicion
            print "\t Surroundings: "
            self.ver(mundo, show = True)
```

### Azucar

```python
class Azucar:
    """
    Montoncitos de azucar de tamaños variables en una  posición especifica en una cuadricula discreta.
    """
    
    def __init__(self, id ,tipo = 1, x = None, y = None, cantidad = None):
        self.id = id
        self.tipo = tipo
        if x == None or y == None:
            self.posicion = np.array([randint(0,50), randint(0,50)])
        else:
            self.posicion = np.array([x,y])
        self.existence = False
        if cantidad == None:
            self.cantidad = randint(10,100)
        else: 
            self.cantidad = cantidad
```

### Mundo

```python
class Mundo:
    """
    Un Mundo de SugarScape representado por un array de NumPy de 2D de 51x51
    world = NumPy 2D 51x51 array 
    """
    
    def __init__(self):
        """Creamos la cuadricula"""
        self.grid = np.zeros((51,51))
        self.sugars_list = []
    
    def place_sugars(self, n, practice_sugars = None):
        """
        Function that creates n number of sugars and places them both on the grid and the sugars_list.
        n: number of sugars to be created, though you can't specify each one.
        practice_sugars: list that contains handcrafted sugars in specific locations for trials purposes.
        """
        if practice_sugars == None:
            self.sugars_list = [Azucar(i) for i in xrange(n)]
            self.definirse(self.sugars_list)       
        else:
            self.sugars_list = practice_sugars
            self.definirse(self.sugars_list)
                            
    
    def definirse(self, agent_list):
        """
        Les da existencia a los agentes en el mundo
        Es decir, en una cuadricula (NumPy array de 2 dimensiones),
        la función coloca cada agente en las cordenadas aleatorias que tiene cada uno en self.posicion
        Al ser colocado se dice que el agente existe
        Si esa casilla ya esta ocupada, se buscan nuevas coordenadas aleatorias para existir hasta que se logre. 
        """
        for i in xrange(len(agent_list)):
            while agent_list[i].existence == False:
                x,y = agent_list[i].posicion
                if self.grid[x,y] == 0:
                    self.grid[x,y] = agent_list[i].tipo
                    agent_list[i].existence = True
                else:
                    agent_list[i].posicion = np.array([randint(0,50), randint(0,50)])  
            
    def actualizar_agentes(self, agentes):
        for i in xrange(len(agentes)):
            self.grid[agentes[i].posicion[0], agentes[i].posicion[1]] = 0
            agentes[i].moverse(self)
            self.grid[agentes[i].posicion[0], agentes[i].posicion[1]] = agentes[i].tipo
        
    def dibujarse(self):
        plt.figure(figsize=(11,10), dpi=80)
        pyplot.pcolor(-numpy.flipud(self.grid))
        plt.axis([0, 51, 0, 51])
        plt.gray()
        plt.colorbar() #ELIMINAR es para fines de pruebas
#         plt.axis('off') #ELIMINAR
```
