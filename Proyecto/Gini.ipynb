# Coeficiente de Gini

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lorenz(incomes):
    n = len(incomes)
    c = sort(incomes)
    c = cumsum(c)
    c = np.insert(c,[0],0)
    points = np.zeros([n+1,2])
    for i in xrange(n+1):
        points[i,0] = 100*(i+0.0)/n
        points[i,1] = 100*(c[i]+0.0)/c[n]
    return points

def gini(incomes):
    n = len(incomes)
    S = sum(lorenz(incomes)[:,1])
    return  (100 + (100 - 2*S)/(n + 0.0))   
```

```python
incomes_1 = [10,10,500, 10,10,20,20,52,5,85,9,5,9,5,9,445,458,10,10,1000, 10,10,10,10,20,20,30,40,50,60,70,100]
Points = lorenz(incomes_1)
X = Points[:,0]
Y = Points[:,1]
```
