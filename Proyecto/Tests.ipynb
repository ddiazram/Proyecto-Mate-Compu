# Test 2

Random Agents

```python
SS = World()
agents = [(Agent_grid(i)) for i in xrange(100)]
SS.define(agents)
SS.place_sugars(n = 500)
SS.state(agents)
SS.gini(agents, plot = True)
SS.draw()

steps=20

for i in xrange(1,steps+1): 
    SS.update_agents(agents)
    SS.update_sugars()
    SS.state(agents)
    SS.gini(agents, plot = True)
    SS.draw()
```

# Test 3

Diferent kinds of death an agent

```python
SS_death = World()
doomed_energy = Agent_grid(id = 0, x = 25, y = 25, metabolism = 3, energy= 10, vision=2)
doomed_life = Agent_grid(id = 1, x = 20, y = 20, age = 5, max_age=9, energy= 1000)
agents = [doomed_energy, doomed_life]
SS_death.define(agents)
SS_death.state(agents)
doomed_energy.state(SS, show_existance=True)
doomed_life.state(SS, show_existance=True)
SS_death.draw()

steps = 5
for i in xrange(steps):
    SS_death.update_agents(agents)
    SS_death.state(agents)
    doomed_energy.state(SS, show_existance=True)
    doomed_life.state(SS, show_existance=True)
    SS_death.draw()

```

# Test 4: Reproducction

2 Steps in wich an agent is very likely to reporduce

```python
SS =  World()
asexual_agent = Agent_grid(0, position=[25,25], energy= 1000, vision = 2)
agents = [asexual_agent]
SS.define(agents)
asexual_agent.state(SS, show_existence= True)
SS.state(agents)
SS.draw()

SS.update_agents(agents)
for agent in agents:
    agent.state(SS, show_existence= True)  
SS.state(agents)
SS.draw()

SS.update_agents(agents)
for agent in agents:
    agent.state(SS, show_existence= True)  
SS.state(agents)
SS.draw()
```
